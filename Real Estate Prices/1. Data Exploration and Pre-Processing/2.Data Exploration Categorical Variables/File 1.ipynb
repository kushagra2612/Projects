```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
td  = pd.read_csv('Transformed_Housing_Data.csv')
```

```python
td.head()
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
td  = td.rename(columns = {'Sale Price' : 'Sale_Price'})
```

```python
td  = td.rename(columns = {'Condition of the House' : 'Condition_of_the_House'})
```

```python
td  = td.rename(columns = {'Ever Renovated' : 'Ever_Renovated'})
```

```python
td  = td.rename(columns = {'Waterfront View' : 'Waterfront_View'})
```

```python
td.head()
```

```python
td['Waterfront_View'].unique()
```

```python
mod = ols('Sale_Price ~ Condition_of_the_House', data  = td).fit()
```

```python
Anova_Table = sm.stats.anova_lm( mod, typ = 2)
```

```python
print(Anova_Table)
```

```python
mod = ols('Sale_Price ~ Waterfront_View', data  = td).fit()
sm.stats.anova_lm( mod, typ = 2)
```

```python
mod = ols('Sale_Price ~ Ever_Renovated', data  = td).fit()
sm.stats.anova_lm( mod, typ = 2)
```

```python
mod = ols('Sale_Price ~ Zipcode', data  = td).fit()
sm.stats.anova_lm( mod, typ = 2)
```

```python
td = pd.get_dummies(td, columns = ['Condition_of_the_House'], drop_first = True)
```

```python
td.head()
```

```python
td = pd.get_dummies(td, columns = ['Ever_Renovated', 'Waterfront_View'], drop_first = True)
```

```python
td.head()
```

```python
# Zip_Table
ZT = td.groupby('Zipcode').agg({'Sale_Price' : 'mean'}).sort_values('Sale_Price', ascending = True)
```

```python
ZT.head()
```

```python
ZT['Zipcode_Group'] = pd.cut(ZT['Sale_Price'],bins = 10, 
                             labels = ['Zipcode_Group_0',
                                      'Zipcode_Group_1',
                                      'Zipcode_Group_2',
                                      'Zipcode_Group_3',
                                      'Zipcode_Group_4',
                                      'Zipcode_Group_5',
                                      'Zipcode_Group_6',
                                      'Zipcode_Group_7',
                                      'Zipcode_Group_8',
                                      'Zipcode_Group_9'],
                            include_lowest = True)
```

```python
ZT = ZT.drop(columns = "Sale_Price")
```

```python
td = pd.merge(td,ZT,left_on = 'Zipcode', how = 'left', right_index = True)
```

```python
td = td.drop(columns = "Zipcode")
```

```python
td.head()
```

```python
td = pd.get_dummies(td, columns = ['Zipcode_Group'], drop_first = True)
```
