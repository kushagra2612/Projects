```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action = 'ignore')
```

```python
data = pd.read_csv("Transformed_Housing_Data2.csv")
data.head()
```

```python
data['No of Times Visited'] = np.where(data['No of Times Visited']==0, 
                                       np.nan ,data['No of Times Visited'] )
```

```python
data['No of Times Visited'].head()
```

```python
data['No of Times Visited'].isnull().sum()/len(data)
```

```python
data['Years Since Renovation'].value_counts()
```

```python
data['Years Since Renovation'] = np.where(data['Years Since Renovation']==0 ,
                                           np.nan, data['Years Since Renovation'])
data['Years Since Renovation'].isnull().sum()
```

```python
data.head()
```

```python
data['Basement Area (in Sqft)'] = np.where( data['Basement Area (in Sqft)'] ==0 ,
                                           np.nan, data['Basement Area (in Sqft)'])
data['Basement Area (in Sqft)'].isnull().sum()
```

```python
data.head()
```

```python
data.to_csv("NaN_Housing_Prices.csv", index = False)
```

```python
data.isnull().sum()
```

```python
for i in data.columns:
  tmp = data[i].isnull().sum()/len(data)
  if tmp > 0.70:
    print("Removing column '"+i+"' with missing value ratio", tmp)
    data = data.drop(columns = [i])
```
